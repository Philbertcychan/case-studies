## üí∞ Monetization Levers

- **Token-based pricing**: Default, flexible
- **Seat-based pricing**: For hosted dashboards / RAG pipelines
- **Feature gating**: Fine-tuning, latency SLAs, deployment control
- **Support tiers**: Community vs Dedicated
- **Region-based pricing**: Discounted rates for emerging markets

This hybrid model supports flexibility while allowing premium capture.
