## üåç Section 1: Segment Identification & Market Mapping

In this section, we define high-priority customer segments for Cohere based on real-world applications of large language models (LLMs).

### üîé Selection Criteria:
We identify segments based on:
- Size of market opportunity (TAM)
- Readiness to adopt LLM technology
- Alignment with Cohere's strengths (privacy, multilingual support, enterprise focus)

---

### üß© Segments Defined

| Segment                  | Description                                                             |
|--------------------------|-------------------------------------------------------------------------|
| Enterprise Search        | Internal Q&A + document search tools for employees                     |
| Customer Support (BPO)   | AI-driven customer service & call center automation                    |
| Legal & Compliance       | Document analysis, contract review, regulatory compliance               |
| Healthcare NLP           | Patient notes, summarization, diagnostics                              |
| Government / Sovereign   | National AI efforts, defense, and regulated infrastructure              |


```python
import pandas as pd

# Define segment characteristics
segments = pd.DataFrame({
    'Segment': [
        'Enterprise Search', 
        'Customer Support (BPO)', 
        'Legal & Compliance', 
        'Healthcare NLP', 
        'Government / Sovereign'
    ],
    'Est_TAM_Bn': [30, 80, 20, 50, 40],
    'Adoption_Stage': ['Mature', 'Mature', 'Emerging', 'Emerging', 'Early'],
    'Cohere_Strength_Alignment': ['High', 'Medium', 'High', 'Medium', 'High']
})

segments
```

```python
# Convert qualitative to numeric scores
adoption_map = {'Early': 1, 'Emerging': 2, 'Mature': 3}
alignment_map = {'Low': 1, 'Medium': 2, 'High': 3}

segments['Adoption_Score'] = segments['Adoption_Stage'].map(adoption_map)
segments['Alignment_Score'] = segments['Cohere_Strength_Alignment'].map(alignment_map)

# Weighted Score (feel free to tweak weights)
segments['Composite_Score'] = (
    0.5 * segments['Est_TAM_Bn'] + 
    10 * segments['Adoption_Score'] + 
    10 * segments['Alignment_Score']
)

# Sort and visualize
segments = segments.sort_values(by='Composite_Score', ascending=False)
segments[['Segment', 'Est_TAM_Bn', 'Adoption_Score', 'Alignment_Score', 'Composite_Score']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Composite_Score', y='Segment', data=segments, palette='viridis')
plt.title('LLM Segment Prioritization (Composite Score)')
plt.xlabel('Composite Score')
plt.ylabel('Customer Segment')
plt.show()
```

## üåç Section 2: Regional Analysis (Global GTM Priorities)

In this section, we analyze which countries offer the highest strategic potential for Cohere to expand into, based on:

- **AI Readiness**: National infrastructure and talent capacity to adopt AI
- **Economic Potential**: GDP as a proxy for commercial opportunity
- **Regulatory Environment**: Degree of national focus on sovereign AI or data privacy
- **Cohere Fit**: Alignment with Cohere‚Äôs values (privacy-first, multilingual, enterprise focus)

---

### üß™ Methodology

We‚Äôll use a combination of:
- **Oxford Insights AI Readiness Index (2022)** ‚Äî measures national AI capabilities
- **World Bank GDP data (2023)** ‚Äî economic weight of each country
- **Custom Criteria** ‚Äî such as support for sovereign AI, interest in multilingual models, etc.

Each region will be scored on a weighted composite index to prioritize strategic bets.


```python
import pandas as pd

# Simulated dataset: Top countries by AI readiness, GDP, etc.
regions = pd.DataFrame({
    'Country': ['Canada', 'Germany', 'United Arab Emirates', 'United Kingdom', 'France', 'Japan'],
    'AI_Readiness_Score': [81.2, 85.0, 78.4, 86.1, 83.3, 84.5],
    'GDP_Bn_USD': [2263, 4230, 501, 3270, 3187, 4200],
    'Sovereign_AI_Focus': ['High', 'Medium', 'High', 'Low', 'Medium', 'Low'],
    'Multilingual_Advantage': ['High', 'Medium', 'Medium', 'Medium', 'High', 'Low']
})
```

```python
# Map qualitative to scores
sovereign_map = {'Low': 1, 'Medium': 2, 'High': 3}
multi_map = {'Low': 1, 'Medium': 2, 'High': 3}

regions['Sovereign_Score'] = regions['Sovereign_AI_Focus'].map(sovereign_map)
regions['Multi_Score'] = regions['Multilingual_Advantage'].map(multi_map)

# Normalize GDP and AI readiness for scoring (min-max scaling)
regions['Norm_GDP'] = (regions['GDP_Bn_USD'] - regions['GDP_Bn_USD'].min()) / \
                      (regions['GDP_Bn_USD'].max() - regions['GDP_Bn_USD'].min())

regions['Norm_AI_Readiness'] = (regions['AI_Readiness_Score'] - regions['AI_Readiness_Score'].min()) / \
                               (regions['AI_Readiness_Score'].max() - regions['AI_Readiness_Score'].min())

# Composite Score: 0.3 AI readiness, 0.3 GDP, 0.2 sovereign AI, 0.2 multilingual
regions['Composite_Score'] = (
    0.3 * regions['Norm_AI_Readiness'] + 
    0.3 * regions['Norm_GDP'] + 
    0.2 * regions['Sovereign_Score'] + 
    0.2 * regions['Multi_Score']
)

regions = regions.sort_values(by='Composite_Score', ascending=False)
regions[['Country', 'Composite_Score', 'AI_Readiness_Score', 'GDP_Bn_USD', 'Sovereign_AI_Focus', 'Multilingual_Advantage']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(x='Composite_Score', y='Country', data=regions, palette='magma')
plt.title('Regional GTM Opportunity Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.show()
```

## üéØ Section 3: Opportunity Prioritization Framework

To allocate limited go-to-market resources (sales, partnerships, research focus), Cohere needs a consistent way to evaluate and prioritize opportunities ‚Äî whether enterprise customers, regions, or product lines.

This framework scores opportunities across five key dimensions:

- **Strategic Fit** ‚Äî Does this align with our long-term direction?
- **Revenue Potential** ‚Äî Size and scale of potential deal or market
- **Speed to Value** ‚Äî How quickly can we launch or integrate?
- **Competitive Advantage** ‚Äî Do we have a moat or differentiator?
- **Execution Risk** ‚Äî Internal capability or external friction

---

### üß™ Methodology

Each opportunity is rated 1‚Äì5 on each dimension, and weighted to generate a composite priority score.

This helps us objectively sort through large lists and identify Tier 1, 2, 3 opportunities for follow-up.

We'll simulate a scoring of 5 hypothetical opportunities:
- **Canadian Gov Cloud Deal**
- **Partnership with Shopify**
- **Expansion into UAE**
- **Enterprise Sales into German Banks**
- **French Translation API for Education**


```python
import pandas as pd

# Simulated scoring matrix
data = {
    'Opportunity': [
        'Canadian Gov Cloud Deal',
        'Shopify Partnership',
        'UAE Expansion',
        'German Banks',
        'French Education API'
    ],
    'Strategic_Fit': [5, 4, 3, 4, 3],
    'Revenue_Potential': [4, 5, 3, 5, 3],
    'Speed_to_Value': [2, 5, 4, 3, 4],
    'Competitive_Advantage': [5, 4, 4, 3, 5],
    'Execution_Risk': [2, 3, 3, 4, 2]  # lower = better
}

df = pd.DataFrame(data)
```

```python
# Invert Execution Risk to turn it into a 'score' (lower risk = higher score)
df['Risk_Score'] = 6 - df['Execution_Risk']

# Composite Score: weighted sum
# Weights: Strategic Fit (25%), Revenue (25%), Speed (15%), Advantage (20%), Risk (15%)
df['Composite_Score'] = (
    0.25 * df['Strategic_Fit'] +
    0.25 * df['Revenue_Potential'] +
    0.15 * df['Speed_to_Value'] +
    0.20 * df['Competitive_Advantage'] +
    0.15 * df['Risk_Score']
)

df = df.sort_values(by='Composite_Score', ascending=False)
df[['Opportunity', 'Composite_Score']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x='Composite_Score', y='Opportunity', data=df, palette='viridis')
plt.title('Opportunity Prioritization Score')
plt.xlabel('Score')
plt.ylabel('Opportunity')
plt.tight_layout()
plt.show()
```

## üí∞ Section 4: Pricing and Profitability Simulation

Cohere, like many API and AI platform businesses, must carefully balance:

- Accessibility to small builders/startups
- Value capture from large enterprise buyers
- Infrastructure cost and margin compression

This notebook simulates a basic tiered pricing structure and evaluates:

- Estimated monthly revenue by tier
- Infrastructure costs per tier
- Gross margin % across different customer distributions
- Upside from moving customers into higher tiers

---

### üìä Pricing Model Assumptions

| Tier        | Monthly Price | Cost per Customer | Avg Monthly Users |
|-------------|----------------|--------------------|-------------------|
| Starter     | $49            | $15                 | 500               |
| Pro         | $249           | $40                 | 300               |
| Enterprise  | $1,499         | $300                | 50                |

---

### üìà Simulation Goals

- Estimate monthly **revenue** and **costs**
- Calculate **gross margin**
- Simulate impact of converting 10% of Pro users to Enterprise


```python
import pandas as pd

data = {
    'Tier': ['Starter', 'Pro', 'Enterprise'],
    'Monthly_Price': [49, 249, 1499],
    'Cost_per_Customer': [15, 40, 300],
    'Monthly_Users': [500, 300, 50]
}

df = pd.DataFrame(data)

# Revenue and Cost
df['Monthly_Revenue'] = df['Monthly_Price'] * df['Monthly_Users']
df['Monthly_Cost'] = df['Cost_per_Customer'] * df['Monthly_Users']
df['Gross_Profit'] = df['Monthly_Revenue'] - df['Monthly_Cost']
df['Gross_Margin_%'] = (df['Gross_Profit'] / df['Monthly_Revenue']) * 100

df
```

```python
# Clone and simulate 10% Pro users upgrading to Enterprise
upgrade_df = df.copy()

pro_users = upgrade_df.loc[upgrade_df['Tier'] == 'Pro', 'Monthly_Users'].values[0]
upgrade_count = int(pro_users * 0.10)

# Update Pro
upgrade_df.loc[upgrade_df['Tier'] == 'Pro', 'Monthly_Users'] -= upgrade_count

# Update Enterprise
upgrade_df.loc[upgrade_df['Tier'] == 'Enterprise', 'Monthly_Users'] += upgrade_count

# Recalculate
upgrade_df['Monthly_Revenue'] = upgrade_df['Monthly_Price'] * upgrade_df['Monthly_Users']
upgrade_df['Monthly_Cost'] = upgrade_df['Cost_per_Customer'] * upgrade_df['Monthly_Users']
upgrade_df['Gross_Profit'] = upgrade_df['Monthly_Revenue'] - upgrade_df['Monthly_Cost']
upgrade_df['Gross_Margin_%'] = (upgrade_df['Gross_Profit'] / upgrade_df['Monthly_Revenue']) * 100

upgrade_df
```

```python
import matplotlib.pyplot as plt

# Total metrics
before = df[['Monthly_Revenue', 'Gross_Profit']].sum()
after = upgrade_df[['Monthly_Revenue', 'Gross_Profit']].sum()

metrics = pd.DataFrame({
    'Metric': ['Revenue', 'Gross Profit'],
    'Before': [before['Monthly_Revenue'], before['Gross_Profit']],
    'After': [after['Monthly_Revenue'], after['Gross_Profit']]
})

metrics.set_index('Metric').plot(kind='bar', figsize=(8, 5), colormap='coolwarm')
plt.title('Impact of 10% Pro ‚Üí Enterprise Upgrade')
plt.ylabel('USD')
plt.tight_layout()
plt.show()
```

## üìà Section 5: Valuation Model & Scenario Planning

A core part of strategic planning is estimating a company‚Äôs valuation under various business scenarios. In this section, we:

- Simulate Cohere‚Äôs valuation using common SaaS multiples
- Model 3 scenarios: **Bear, Base, Bull**
- Adjust for differences in revenue growth and margin performance
- Visualize valuation sensitivity to pricing improvements or user growth

---

### üîß Assumptions:

- Revenue is simulated based on Cohere‚Äôs three product tiers
- SaaS revenue multiples range from 8x to 15x
- Gross margin improvements unlock higher valuation in Bull case

---

### üßÆ Base Metrics from Pricing Model

These values come from the `pricing_simulation` notebook:

- Monthly Revenue: ~$300K+
- Gross Margin: ~70%
- ARR (Annual Recurring Revenue): Monthly Revenue √ó 12


```python
# Base assumptions (from pricing simulation)
monthly_revenue = 379700  # From pricing simulation upgrade scenario
gross_margin = 0.74

# Annualize
ARR = monthly_revenue * 12

print(f"Annual Recurring Revenue (ARR): ${ARR:,.2f}")
```

```python
# SaaS valuation multiples
bear_multiple = 8
base_multiple = 12
bull_multiple = 15

# Revenue assumptions
scenarios = {
    'Bear': ARR * 0.85,
    'Base': ARR,
    'Bull': ARR * 1.2
}

# Gross margin sensitivity
gross_margins = {
    'Bear': 0.60,
    'Base': 0.74,
    'Bull': 0.82
}

# Valuation calculation
valuation = {
    name: {
        'ARR': revenue,
        'Gross_Margin': gross_margins[name],
        'Valuation': revenue * multiple
    }
    for name, revenue, multiple in zip(scenarios.keys(), scenarios.values(), [bear_multiple, base_multiple, bull_multiple])
}

val_df = pd.DataFrame(valuation).T
val_df['Valuation'] = val_df['Valuation'].apply(lambda x: round(x / 1e6, 2))  # Convert to millions
val_df
```

```python
val_df['Valuation'].plot(kind='bar', colormap='summer', figsize=(7, 4))
plt.title('Cohere Valuation Estimate by Scenario')
plt.ylabel('Valuation (Millions USD)')
plt.xlabel('Scenario')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
