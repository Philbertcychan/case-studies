```python
import pandas as pd

data = {
    "Provider": ["OpenAI", "Cohere", "Anthropic", "Mistral"],
    "Model": ["GPT-4o", "Command R+", "Claude 3 Sonnet", "Mistral Large"],
    "Input ($/1M tokens)": [5.0, 3.0, 3.0, 2.0],
    "Output ($/1M tokens)": [15.0, 15.0, 15.0, 6.0]
}

df = pd.DataFrame(data)
df["Total ($/1M tokens)"] = df["Input ($/1M tokens)"] + df["Output ($/1M tokens)"]
df
```

```python
import matplotlib.pyplot as plt

# Sort by total cost
df_sorted = df.sort_values("Total ($/1M tokens)", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(df_sorted["Provider"], df_sorted["Total ($/1M tokens)"], color="steelblue")
plt.xlabel("Total Cost per 1M Tokens ($)")
plt.title("LLM Provider Pricing Comparison")
plt.gca().invert_yaxis()  # Highest at top
plt.grid(True, axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.savefig("../report/graphics/pricing_comparison.png", dpi=300)
plt.show()
```
